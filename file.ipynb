```python
import pandas as pd

df = pd.read_csv('new3.csv')

df.head()

df.info()

df.describe()
df.isnull().sum()

df.columns
```

```python
# Normalize all the columns used
df['elevation_modified'] = df['elevation']
df.loc[df['elevation'] < 8000, 'elevation_modified'] = df['elevation'] * 0.5
df['elevation_modified'] = df['elevation_modified'] / df['elevation_modified'].max()

df['population_sq_km_modified'] = df['Population per sq. km.'] / df['Population per sq. km.'].max()

df['temp_modified'] = df['temp (mean)'] / df['temp (mean)'].max()

df['border_modified'] = df['border'] / df['border'].max()

df['terrorism_modified'] = df['terrorist attacks'] / df['terrorist attacks'].max()

df['maoist_modified'] = df['maoist attacks'] / df['maoist attacks'].max()
```

```python
print(df.head())
```

```python
import numpy as np

elevation_modified_coef_risk = 1
border_modified_coef_risk = 1
terrorism_modified_coef_risk = 1
maoist_modified_coef_risk = 1
elevation_modified_coef_hardship = 1
LWE_areas_coef_hardship = 1
temp_mean_coef_hardship = 1
population_sq_km_coef_hardship = 1

df['Risk'] = ( elevation_modified_coef_risk* df['elevation_modified'] + border_modified_coef_risk* df['border_modified'] + terrorism_modified_coef_risk*df['terrorism_modified'] + maoist_modified_coef_risk*df['maoist_modified'])
df['Hardship'] = (elevation_modified_coef_hardship* df['elevation_modified'] + LWE_areas_coef_hardship*df['LWE areas'] + temp_mean_coef_hardship*df['temp_modified'] - population_sq_km_coef_hardship* df['population_sq_km_modified'])

max_value = max(df['Risk'].max(), df['Hardship'].max())
min_value = min(df['Risk'].min(), df['Hardship'].min())
df['Risk'] = (df['Risk'] - min_value) / (max_value - min_value)
df['Hardship'] = (df['Hardship'] - min_value) / (max_value - min_value)

# Create the 3x3 matrix
risk_hardship_matrix = np.array([
    [df['Risk'][0] * df['Hardship'][0], df['Risk'][0] * df['Hardship'][1], df['Risk'][0] * df['Hardship'][2]],
    [df['Risk'][1] * df['Hardship'][0], df['Risk'][1] * df['Hardship'][1], df['Risk'][1] * df['Hardship'][2]],
    [df['Risk'][2] * df['Hardship'][0], df['Risk'][2] * df['Hardship'][1], df['Risk'][2] * df['Hardship'][2]]
])
risk_hardship_matrix
```

```python
# Calculate the risk_hardship_matrix_index 
df['risk_hardship_matrix_index'] = df['Risk'] * df['Hardship']

risk_hardship_matrix = np.outer(df['Risk'], df['Hardship'])

df.head()
```

```python
df['risk_value'] = 0
df.loc[df['Risk'] < 0.333333, 'risk_value'] = 1
df.loc[(df['Risk'] >= 0.333333) & (df['Risk'] < 0.666666), 'risk_value'] = 2
df.loc[df['Risk'] >= 0.666666, 'risk_value'] = 3


df['hardship_value'] = 0
df.loc[df['Hardship'] < 0.333333, 'hardship_value'] = 1
df.loc[(df['Hardship'] >= 0.333333) & (df['Hardship'] < 0.666666), 'hardship_value'] = 2
df.loc[df['Hardship'] >= 0.666666, 'hardship_value'] = 3
print(df.head())
```

```python
df['risk_hardship_matrix_index'] = df['risk_value'] * df['hardship_value']
risk_hardship_matrix = np.outer(df['risk_value'], df['hardship_value'])
print(risk_hardship_matrix)
```
